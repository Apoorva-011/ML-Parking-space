```python
import pandas as pd
import seaborn as sns
import numpy as np
data=pd.read_csv("Parking_lots_dataset.csv")

```

```python
numerical_cols = ["Capacity","Occupancy"]
for col in numerical_cols:
    data[col].fillna(data[col].mean(), inplace=True)
```

```python
data.describe
```

```python
data=data.drop_duplicates()
```

```python
data.describe()
```

```python
data.shape
```

```python
data = pd.get_dummies(data,columns=['Parking_Number'], prefix='')
```

```python
from sklearn.model_selection import train_test_split
x=data.drop(['Occupancy'],axis=1)
y=data['Occupancy']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=50)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
xtrain=sc.fit_transform(x_train)
xtest=sc.transform(x_test)

```

```python
from sklearn.linear_model import LinearRegression
regr=LinearRegression()
regr.fit(xtrain,y_train)
regr.score(xtest,y_test)
```
